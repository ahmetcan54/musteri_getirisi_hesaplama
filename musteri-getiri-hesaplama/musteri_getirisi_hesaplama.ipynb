```python
agg_df = df.groupby(["COUNTRY","SOURCE","SEX","AGE"])["PRICE"].mean() # burada ülkelerin cihazlarına cihazların cinsiyete cinsiyetin ise yaşa göre fiyat ortalaması alındı.
```

```python
agg_df.sort_values( ascending= False,inplace=True) # burada ise fiyat değişkeni büyükten küçüğe sıralanarak güncellendi
```

```python
agg_df = agg_df.reset_index() # burada ise index olan isimleri değişkene dönüştürüldü.
```

```python
agg_df.head() # burada göründüğü gibi.
```

```python
agg_df["AGE_CAT"] = pd.cut(
    agg_df["AGE"],
    bins=[0, 18, 24, 31, 41, 70],
    labels=["0_18","19_23","24_30","31_40","41_70"]
) # burada yaş değişkenini  cut fonksiyonu ile kategorik hale getiriyoruz önemli !!!
```

```python
agg_df.head()
```

```python
agg_df["customers_level_based"] = [row[0].upper() + "_" + row[1].upper() + "_" + row[2].upper() + "_" + row[5].upper() for row in agg_df.values]
# burada ise aşağıdaki gibi değişkenlerimizi birleştirdk.
```

```python
agg_df.head()
```

```python
agg_df = agg_df[["customers_level_based","PRICE"]] # HEPSİNİ BİR ARAYA GETİRDİĞİMİZ İÇİN DİĞER DEĞİŞKENLERİ DROPLADIK.
agg_df.head()
```

```python
agg_df["customers_level_based"].value_counts() # burada değerlerden birden fazla olduğundan bunların ortalamalarını alarak tek hale getireceğiz.
```

```python
agg_df = agg_df.groupby("customers_level_based")["PRICE"].mean() # burada yukarıdaki işlemi yaptık ve sorun çözüldü.
```

```python
agg_df.head()
```

```python
agg_df = agg_df.reset_index() # burada index olan değeri değişkene çeviriyoruz.
```

```python
agg_df.head()
```

```python
agg_df["customers_level_based"].value_counts() # kontrol amaçlı.
```

```python
agg_df["SEGMENT"] = pd.qcut(agg_df["PRICE"], 4, labels=["D", "C", "B", "A"])
agg_df.head(30)
agg_df.groupby("SEGMENT").agg({"PRICE": "mean"}) # burada değerlerimizi segmentlere ayırdık.
```

```python
agg_df.head()
```
